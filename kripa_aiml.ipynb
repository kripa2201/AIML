```python
#fill null values with mean value
for i in df.columns:# CHECKING FOR NULL VALUE IN EVERY COLUMN
    if df[i].isnull().sum()>0:
        df[i].fillna(df[i].mean(),inplace=TRUE)# REPLACE NULL VALUE WITH MEAN USING FILLNA FUNCTION
```
